```python
# Correlação
import pandas as pd
import numpy as np

ENDERECO_DADOS = 'https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv'

try:
    print('Obtendo dados')
    # Recebendo dados do arquivo CSV
    df_ocorrencias = pd.read_csv(ENDERECO_DADOS, sep=';', encoding='iso-8859-1')

    # Delimitando as variáveis para CISP, lesao_corp_dolosa e lesao_corp_morte
    df_lesoes = df_ocorrencias[['cisp', 'lesao_corp_dolosa','lesao_corp_morte']]

    # Agrupando por CISP e somando as ocorrências de lesão corporal dolosa e morte
    df_total_lesoes = df_lesoes.groupby('cisp', as_index=False)[['lesao_corp_dolosa', 'lesao_corp_morte']].sum()

    print(df_total_lesoes.head(15))

except Exception as e:
    print(f'Erro ao obter dados: {e}')
```
